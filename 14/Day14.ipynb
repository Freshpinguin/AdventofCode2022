```python
with open('input14.txt', 'r') as this:
    inp = this.read().split('\n')[:-1]
```

```python
def line_to_coord(line: str) -> [(int,int)]:
    return [eval(x) for x in line.split(' -> ')]
```

```python
def coord_between(coord1: (int,int), coord2: (int,int)) -> [(int,int)]: 
    if coord1[0] == coord2[0]:
        return [(coord1[0],i) for i in range(min(coord1[1],coord2[1]),max(coord1[1],coord2[1])+1)]
    
    if coord1[1] == coord2[1]:
        return [(i,coord1[1]) for i in range(min(coord1[0],coord2[0]),max(coord1[0],coord2[0])+1)]
```

```python
def coord_between_list(lis: [(int,int)]) -> [(int,int)]:
    return [y for x in [coord_between(lis[i],lis[i+1]) for i in range(len(lis)-1)] for y in x]
```

```python
def stone_coordinates(inp: [str]) -> [(int,int)]:
    stones = [y for x in [coord_between_list(line_to_coord(line)) for line in inp] for y in x]
    unique_stones = []
    
    for stone in stones:
        if not stone in unique_stones:
            unique_stones += [stone]
            
    return unique_stones
```

```python
def drop_sand(sand_coord: (int,int), stones: [(int,int)] , fixed_sand: [(int,int)]) -> (int,int):
    
    biggest = max([x[1] for x in stones])
    
    if sand_coord[1] > biggest:
        return []
    
    fixed = stones + fixed_sand
    if not (sand_coord[0],sand_coord[1]+1) in fixed:
        return drop_sand((sand_coord[0],sand_coord[1]+1), stones, fixed_sand)
    if not (sand_coord[0]-1,sand_coord[1]+1) in fixed:
        return drop_sand((sand_coord[0]-1,sand_coord[1]+1), stones, fixed_sand)

    
    if not (sand_coord[0]+1,sand_coord[1]+1) in fixed:
        return drop_sand((sand_coord[0]+1,sand_coord[1]+1), stones, fixed_sand)

    
    return sand_coord
    
```

```python
example = """498,4 -> 498,6 -> 496,6
503,4 -> 502,4 -> 502,9 -> 494,9""".split('\n')

stones = stone_coordinates(inp)
fixed_sand = []
cont = True

while(cont):
    #print(fixed_sand)
    new_ = drop_sand((500,0), stones, fixed_sand)
    if not new_:
        break
        
    fixed_sand += [new_]
    
print(len(fixed_sand))
```
